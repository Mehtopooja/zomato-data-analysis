```python
# Cleaning Rate column
```

```python
# Removing "New","-","/5" from rate column
```

```python
import numpy as np

def handlerate(value):
    if value == 'NEW' or value == '-':
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

# Apply the handlerate function to the 'rate' column
df['rate'] = df['rate'].apply(handlerate)

# Display the first few rows of the 'rate' column
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
#finding null values in rate column with mean
```

```python
df.shape
```

```python
# Fill missing values in the 'rate' column with the mean of non-missing values
df['rate'].fillna(df['rate'].mean(), inplace=True)

# Check the count of missing values in the 'rate' column
missing_values_count = df['rate'].isnull().sum()

print("Count of missing values in 'rate' column:", missing_values_count)
```

```python
# Dropping Null Values
```

```python
df.dropna(inplace= True)
df.head()
```

```python
#  Listed in(city) and location, both are there, lets keep only one.
```

```python
df = df.drop(['listed_in(city)'], errors='ignore')
```

```python
df['cost2Plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else: 
        return float (value)
df['Cost2Plates'] =df['cost2Plates'].apply(handlecomma)
df['cost2Plates'].unique()
```

```python
df.head()
```

```python
# cleaning Rest Type Column 
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]

def handle_location(value):
     if(value in location_lessthan300):
            return 'others'
    else:
            return value
        
df['locatiion']= df['location'].apply(handle_location)
df[location].value_counts()
```

```python
location_counts = df['location'].value_counts()

# Identify locations with less than 300 occurrences
locations_to_replace = location_counts[location_counts < 300].index

# Label those locations as 'others'
df['location'] = df['location'].apply(lambda x: 'others' if x in locations_to_replace else x)

# Display the value counts of the 'location' column
location_counts = df['location'].value_counts()
print(location_counts)
```

```python
df.head()
```

```python
cuisine_counts = df['cuisines'].value_counts()

# Identify cuisines with less than 300 occurrences
cuisines_to_replace = cuisine_counts[cuisine_counts < 100].index

# Label those cuisines as 'others'
df['cuisines'] = df['cuisines'].apply(lambda x: 'others' if x in cuisines_to_replace else x)

# Display the value counts of the 'cuisines' column
cuisine_counts = df['cuisines'].value_counts()
print(cuisine_counts)
```
